```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Mall_Customers.csv")

is_male = df["is_male"].values
income = df["annual-income"].values
score = df["spending-score"].values
age = df["age"].values
```

## Whole dataset

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.set_xlabel('annual-income')
ax.set_ylabel('spending-score')
ax.set_zlabel('age')

colors = np.full(is_male.shape, "blue", dtype=object)
colors[is_male == 0] = "crimson"

ax.scatter(income, score, age, color=colors)
ax.set_title("Whole dataset")

plt.show()
```

## Score X Income

```python
figure, ax = plt.subplots()
ax.scatter(income, score)

ax.set_ylabel("spending score")
ax.set_xlabel("annual income (kU$)")

plt.show()
```

```python
figure, ax = plt.subplots()
ax.bar(is_male, score)

ax.set_ylabel("spending score")
ax.set_xlabel("is_male")

plt.show()
```
